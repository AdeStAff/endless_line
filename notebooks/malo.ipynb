## Missing data

```python
def check_missing_data(df, name):
    print(f"Missing data in {name}:")
    print(df.isnull().sum())  # Count missing values in each column
    print("\nPercentage of missing data:")
    print((df.isnull().sum() / len(df)) * 100)  # Percentage of missing values
    print("-" * 50)
```

```python

check_missing_data(weather, "weather_data")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
weather['dt_iso'] = weather['dt_iso'].astype(str)
weather['dt_iso'] = weather['dt_iso'].str.replace(r' \+0000 UTC$', '', regex=True)
# Convert the 'dt_iso' column to datetime
weather['dt_iso'] = pd.to_datetime(weather['dt_iso'])

# Filter the data for the year 2022
weather_2022 = weather[weather['dt_iso'].dt.year == 2022]

# Plot the 'temp' column for the year 2022
plt.figure(figsize=(10, 6))
plt.plot(weather_2022['dt_iso'], weather_2022['temp'], marker='o', linestyle='-', color='b')
plt.title('Temperature in 2022')
plt.xlabel('Date')
plt.ylabel('Temperature')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Filter the data for the month of July 2022
weather_july_2022 = weather_2022[weather_2022['dt_iso'].dt.month == 4]

# Plot the 'temp' column for July 2022
plt.figure(figsize=(10, 6))
plt.plot(weather_july_2022['dt_iso'], weather_july_2022['temp'], marker='o', linestyle='-', color='r')
plt.title('Temperature in July 2022')
plt.xlabel('Date')
plt.ylabel('Temperature')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
avg_temp_by_id = weather.groupby('weather_id')['temp'].mean()
print(avg_temp_by_id)
```

```python
avg_temp_by_id = weather.groupby('weather_id')['dew_point'].mean()
print(avg_temp_by_id)
```

```python
# Identify numerical columns
numerical_columns = weather.select_dtypes(include='number').columns

# Apply mean aggregation on numerical columns
numerical_avg = weather.groupby('weather_id')[numerical_columns].mean()

# Identify non-numerical columns
non_numerical_columns = weather.select_dtypes(exclude='number').columns

# For non-numerical columns, we can use aggregation like 'first', 'last', or 'mode' (if there are multiple unique values)
non_numerical_avg = weather.groupby('weather_id')[non_numerical_columns].agg(lambda x: x.mode()[0] if not x.mode().empty else None)

# Combine numerical and non-numerical results
result = pd.concat([numerical_avg, non_numerical_avg], axis=1)

# Display the final aggregated DataFrame
print(result)

```

```python
data = pd.read_csv("data/portaventura_world_data.csv")
```

```python
data
```

```python
weather['dt'] = pd.to_datetime(weather['dt'], unit='s')

# Convert 'USAGE_DATE' in attendance to datetime if it's not already
data['USAGE_DATE'] = pd.to_datetime(data['USAGE_DATE'])

# Group by date and calculate the average for numerical columns
weather_numerical = weather.select_dtypes(include=['float64', 'int64'])
data_numerical = data.select_dtypes(include=['float64', 'int64'])

# Group by date and calculate the average for numerical columns
weather_daily_avg = weather_numerical.groupby(weather['dt'].dt.date).median().reset_index()
data_daily_avg = data_numerical.groupby(data['USAGE_DATE'].dt.date).median().reset_index()
```

```python
merged_df = pd.merge(weather_daily_avg, data_daily_avg, left_on='dt', right_on='USAGE_DATE')
```

```python
numerical_columns = merged_df.select_dtypes(include=['float64', 'int64']).columns

# Drop rows with missing values
merged_df_clean = merged_df.dropna(subset=numerical_columns)

# Now you can calculate correlations or further analysis
correlation_matrix = merged_df_clean[numerical_columns].corr()

# Optionally plot the correlation matrix or perform other tasks
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=False, cmap='coolwarm', fmt='.2f')
plt.title('Correlation between Weather and Attendance Data (Daily Averages)')
plt.show()
```

```python
# Compute the correlation between 'temp' and 'PARK_ATTENDANCE'
correlation_temp_attendance = merged_df_clean['temp'].corr(merged_df_clean['PARK_ATTENDANCE'])

# Print the correlation value
print(f'Correlation between Temperature and Park Attendance: {correlation_temp_attendance}')
```
