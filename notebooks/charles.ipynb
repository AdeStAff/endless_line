```python
# Select only numeric columns and compute correlation
numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns
correlation_matrix = df[numeric_columns].corr()

# Create heatmap (same as above)
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix,
            annot=True,
            cmap='coolwarm',
            center=0,
            fmt='.2f',
            square=True)

plt.title('Correlation Matrix of Numeric Features')
plt.tight_layout()
plt.show()
```

```python
daily_guests = df.groupby('USAGE_DATE')['GUEST_CARRIED'].sum().reset_index()
daily_guests.head(10)
```

```python
comparison_df = pd.merge(
    daily_guests,
    df_attendance[['USAGE_DATE', 'attendance']],
    on='USAGE_DATE',
    how='inner'
)
comparison_df["multiple"] = comparison_df["GUEST_CARRIED"]/comparison_df["attendance"]
comparison_df.head(10)
```

```python
# See who is the worst attraction in waiting time

avg_wait_times = df.groupby(['ENTITY_DESCRIPTION_SHORT', 'USAGE_DATE'])['WAIT_TIME_MAX'].mean().reset_index()
overall_avg_wait = avg_wait_times.groupby('ENTITY_DESCRIPTION_SHORT')['WAIT_TIME_MAX'].agg([
    'mean',
    'max',
    'count'
]).sort_values('mean', ascending=False)
overall_avg_wait.head(10)
```

```python
# Display top 15 attractions with highest average wait times
print("\nTop 15 Attractions by Average Maximum Wait Time:")
print(overall_avg_wait.head(15))

# Visualize using a bar plot
plt.figure(figsize=(15, 6))
overall_avg_wait.head(15)['mean'].plot(kind='bar')
plt.title('Average Maximum Wait Time by Attraction')
plt.xlabel('Attraction')
plt.ylabel('Average Maximum Wait Time (minutes)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.dates as mdates

# Calculate daily average wait times per attraction
avg_wait_times = df.groupby(['USAGE_DATE', 'ENTITY_DESCRIPTION_SHORT'])['WAIT_TIME_MAX'].mean().reset_index()

# Convert USAGE_DATE to datetime if it's not already
avg_wait_times['USAGE_DATE'] = pd.to_datetime(avg_wait_times['USAGE_DATE'])

# Get top N attractions
top_n = 12  # Number of attractions to show
top_attractions = avg_wait_times.groupby('ENTITY_DESCRIPTION_SHORT')['WAIT_TIME_MAX'].mean()\
                              .sort_values(ascending=False)\
                              .head(top_n).index

# Create one large figure with stacked subplots
fig, axes = plt.subplots(top_n, 1, figsize=(15, 5*top_n))  # One column, multiple rows

for idx, attraction in enumerate(top_attractions):
    attraction_data = avg_wait_times[avg_wait_times['ENTITY_DESCRIPTION_SHORT'] == attraction]
    
    # Calculate 7-day rolling average
    smoothed_data = attraction_data.sort_values('USAGE_DATE')\
                                 .set_index('USAGE_DATE')['WAIT_TIME_MAX']\
                                 .rolling(window=7).mean()
    
    # Plot on respective subplot
    axes[idx].plot(smoothed_data.index, smoothed_data.values, linewidth=2)
    axes[idx].set_title(attraction, pad=15)
    axes[idx].set_xlabel('Date')
    axes[idx].set_ylabel('Wait Time (min)')
    
    # Format x-axis to show only months
    axes[idx].xaxis.set_major_locator(mdates.MonthLocator())
    axes[idx].xaxis.set_major_formatter(mdates.DateFormatter('%b %Y'))
    axes[idx].tick_params(axis='x', rotation=45)
    axes[idx].grid(True, alpha=0.3)

plt.suptitle('7-Day Rolling Average Wait Times by Attraction', fontsize=16, y=1.01)
plt.tight_layout()
plt.show()
```
